#Null values imputation

```python
# mode imputation for null values in city column
matches['City'].replace(' ',np.NaN).head()
```

```python
df=matches['City'].value_counts().index[0]
```

```python
df
```

```python
matches['City'].fillna(df,inplace=True)
matches['City'].head()
```

```python
matches.City.isnull().sum()
```

```python
matches.Margin.value_counts()
```

```python
margin_null=matches.loc[matches['Margin'].isnull(),:]
```

```python
margin_null
```

**Observations:**

we can see all this matches are leads to superover so,

In a super over in cricket, the concept of a "margin" being zero means that the result of the match is determined solely based on the number of runs scored in the super over itself, rather than considering the overall match result.

so we can impute null with Zero

```python
matches.Margin.fillna(0,inplace=True)
```

```python
matches.Margin.isnull().sum()
```

```python
matches.isnull().sum()
```

```python
null_methods=matches.loc[matches.method.isnull(),:]
```

```python
null_methods
```

We can simply drop the methods column

```python
matches.drop(['method'],axis=1,inplace=True)
```

```python
matches.isnull().sum()
```

```python
matches[matches['SuperOver'].isnull()]
```

**Observations:**

In the above 4 rows,we can see the null values in three columns- SuperOver, Winning team and Player_of _match.

so,we will drop these 4 rows using their index numbers

```python
matches=matches.dropna()
```

```python
matches.isnull().sum()
```

```python
# Checking null values in balls dataset

balls.isnull().sum()
```

```python
def missing(balls):
    print (round((balls.isnull().sum() * 100/ len(balls)),2).sort_values(ascending=False))

missing(balls)
```
